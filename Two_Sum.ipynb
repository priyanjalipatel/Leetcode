````python
# Function to check whether any pair exists
# whose sum is equal to the given target value
def two_sum(arr, target):
    # Sort the array
    arr.sort()
    print(arr)
    left, right = 0, len(arr) - 1
    count= 0
    result = []

    # Iterate while left pointer is less than right
    while left < right:
        sum = arr[left] + arr[right]
        # Check if the sum matches the target
        if sum == target:

            count += 1
            result.append(count)
            #left += 1
            right -= 1
        elif sum < target: 
            left += 1  # Move left pointer to the right
        else:
            right -= 1 # Move right pointer to the left

    # If no pair is found
    return count
````

## ðŸ§  Why use `right >= left` (and not `right > left`)?

Because after the left duplicates loop, `left` may **move forward** inside the array.
We still want to count the duplicates on the right **until right crosses left**.
The `=` ensures we donâ€™t accidentally skip the last valid element when they meet.
