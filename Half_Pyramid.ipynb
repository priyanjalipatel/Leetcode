Program to Print Right Half Pyramid Pattern (Star Pattern)

Given an integer N, print N rows of right half pyramid pattern. I
In right half pattern of N rows, the first row has 1 star, second row has 2 stars and so on till the Nth row which has N stars. All the stars are left aligned. 

Input: 3
    Output: 
    *
    **
    ***


    Input: 5
    Output: 
    *
    **
    ***
    ****
    *****

```python
def half_pyramid(num):
    for i in range (0, num):
        for j in range(0, i+1):
            print("*",end=" ")
        print("")   
        
```

```python
half_pyramid(5)
```
