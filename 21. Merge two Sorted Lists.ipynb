`ListNode` class is **the definition of a linked list node**.

Let‚Äôs break it down:

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val      # stores the data (integer in this problem)
        self.next = next    # points to the next node in the list (or None if it's the last node)
```

---

### Role of each part

* **`val`**
  Holds the value for that node (e.g., `1`, `2`, `3`).

* **`next`**
  Points to the **next ListNode object**.
  If `next=None`, it means this node is the **end of the list**.

---

### Example

If we do:

```python
n3 = ListNode(4)
n2 = ListNode(2, n3)
n1 = ListNode(1, n2)
```

It creates a linked list:

```
n1 ‚Üí n2 ‚Üí n3
 1    2    4
```

So:

* `n1.val = 1`, `n1.next = n2`
* `n2.val = 2`, `n2.next = n3`
* `n3.val = 4`, `n3.next = None`

---

### Why do we need this in LeetCode?

Because in **linked lists** we don‚Äôt use arrays or Python lists.
Instead, we chain nodes together using `next`.
The problem statement gives you `list1` and `list2` as `ListNode` objects, not arrays.

---

üëâ Quick way to explain in an interview:

> ‚Äú`ListNode` is just the building block of a linked list.
> It stores a value and a pointer to the next node.
> Together, many `ListNode`s form the full linked list.‚Äù

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        dummy = ListNode(-1)
        current = dummy

        while list1 and list2:
            if list1.val < list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next
            current = current.next

        current.next = list1 if list1 else list2
        return dummy.next
````

## 1Ô∏è‚É£ What problem are we solving?

You have **two already sorted linked lists** (`list1` and `list2`).

Your goal:

* Merge them into **one sorted linked list**
* **Reuse existing nodes** (no new nodes except a helper)
* Return the head of the merged list

---

## 2Ô∏è‚É£ Why use a `dummy` node?

```python
dummy = ListNode(-1)
current = dummy
```

### Purpose of `dummy`

* Acts as a **fake starting node**
* Makes edge cases easy (like empty lists)
* You never lose track of the head

‚ö†Ô∏è The value `-1` does **not matter**.
We return `dummy.next`, not `dummy`.

---

## 3Ô∏è‚É£ What does `current` do?

`current` is a **tail pointer**:

* It always points to the **last node of the merged list**
* We keep moving it forward as we attach nodes

---

## 4Ô∏è‚É£ Main loop (core logic)

```python
while list1 and list2:
```

This runs **as long as both lists have nodes**.

---

### üîπ Compare values

```python
if list1.val < list2.val:
```

* Since lists are sorted, the **smaller value must come next**
* We attach that node to the merged list

---

### üîπ Attach a node

```python
current.next = list1
list1 = list1.next
```

What happens:

* `current.next` points to the chosen node
* Move `list1` forward (because that node is now used)

Same logic applies when `list2` is smaller.

---

### üîπ Move `current`

```python
current = current.next
```

This advances the tail of the merged list.

---

## 5Ô∏è‚É£ What if one list ends?

After the loop:

* One list is empty
* The other may still have nodes

```python
current.next = list1 if list1 else list2
```

Why this works:

* Remaining nodes are **already sorted**
* We can safely attach them as-is

---

## 6Ô∏è‚É£ Final return

```python
return dummy.next
```

Why not `dummy`?

* `dummy` is fake
* `dummy.next` is the **real head** of the merged list
