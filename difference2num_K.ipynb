```python

    def countKDifference(nums, k):
        map = {}
        count = 0
        for i, val in enumerate(nums):
            diff1 = val - k
            diff2 = val + k
            
            if diff1 in map:
                count = count + map[diff1]
                print(diff1)
                print(count)
                print(map[diff1])  
            if diff2 in map:
                count = count + map[diff2]
                print(diff2)
                print(count)
                print(map[diff2])
            
            
            map[val]= map.get(val,0)+1 # Frequency
        return count
```

```python
nums = [11,12,12,11]
k  = 1
```

```python
countKDifference(nums, k)
```

This shows:

When you saw the first 12, it matched one 11 → count = 1.

When you saw the second 12, again matched one 11 → count = 2.

When you saw the last 11, it matched two 12s (frequency 2) → count = 4.
