Given the roots of two binary trees p and q, write a function to check if they are the same or not.

Two binary trees are considered the same if they are structurally identical, and the nodes have the same value.

Time complexity: O(N)
Space complexity: O(log(N))

```python
class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        if not p and not q:
            return True
        if not p or not q or p.val != q.val:
            return False
        
        return(self.isSameTree(p.left,q.left) and
        self.isSameTree(p.right,q.right))
```

The "Same Tree" problem can be solved using
either Depth-First Search (DFS) or Breadth-First Search (BFS). Both methods are valid, have the same time complexity of O(n) (where n is the number of nodes), and can be implemented to correctly check if two binary trees are structurally identical and have the same node values. 
DFS Approach
The DFS approach is generally more intuitive and often implemented recursively. It involves simultaneously traversing both trees in the same order (e.g., pre-order, in-order, or post-order) and comparing the nodes at each corresponding step. 
The logic at each step is:

    If both current nodes are null, they match, so continue.
    If one node is null and the other is not, or if their values are different, the trees are not the same, so return false.
    If they match, recursively check their left subtrees and right subtrees. The final result is true only if both recursive calls return true
