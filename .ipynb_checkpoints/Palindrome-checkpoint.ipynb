```python
def isPalindrome(x):
    str_rev =  '' 
    index = len(x)
    while index>0:
        str_rev += x[index-1]
        index -=1
    if str_rev == x:
        return True
    else:
        return False
```

```python
string = 'reve'
isPalindrome(string)
```

```python
def isPalindrome(x):
    str_rev =  '' 
    s = str(x)
    index = len(s)
    while index>0:
        str_rev += s[index-1]
        index -=1
    if str_rev == s:
        return True
    else:
        return False
```

```python
num = 1214
isPalindrome(num)
```
