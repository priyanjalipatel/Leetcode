```python
height = [1,8,6,2,5,4,8,3,7]
```

```python
def maxArea(height):
    maxA = 0
    for i in range(0, len(height)):
        for j in range (i+1, len(height)):
            w = j- i
            h = min(height[i],height[j])
            area = w * h
            maxA = max(area,maxA)
    return maxA
```

```python
height = [1,8,6,2,5,4,8,3,7]
maxArea(height)
```

```python
def maxArea(height):
    left = 0
    right = len(height) - 1
    maxA = 0

    while left < right:
        w = right - left
        h = min(height[left], height[right])
        area = w * h
        maxA = max(maxA, area)

        # move the smaller height pointer
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return maxA
```

```python
height = [1,8,6,2,5,4,8,3,7]
maxArea(height)
```
