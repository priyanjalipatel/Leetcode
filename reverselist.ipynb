````python
def reverseList(head):
        prev = None
        current = head
        while current is not None:
            next = current.next
            current.next = prev
            
            prev = current
            current = next
            print(prev)
            
        return prev
````

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
````

````python
def build_linked_list(arr):
    dummy = ListNode()
    tail = dummy
    for val in arr:
        tail.next = ListNode(val)
        tail = tail.next
    return dummy.next
````

````python
head = build_linked_list([1,2,3,4,5])
new_head = reverseList(head)
````

---

## Short answer first

`reverseList` returns **`prev`** because:

> **`prev` becomes the NEW HEAD of the reversed linked list**

LeetCode prints the reversed list because **LeetCode uses the returned node as the head** and then traverses it.

---

## What is happening step-by-step

### Original list

```
head → 1 → 2 → 3 → 4 → 5 → None
```

Initial values:

```
prev = None
current = head (1)
```

---

## One loop iteration (important)

At each step:

1. Save next node
2. Reverse the pointer
3. Move `prev` and `current` forward

### After first iteration

```
None ← 1   2 → 3 → 4 → 5 → None
       ↑
     prev
current → 2
```

---

### After full loop finishes

```
None ← 1 ← 2 ← 3 ← 4 ← 5
                     ↑
                   prev
current = None
```

At this point:

* `current` is `None` (end reached)
* `prev` is pointing to **5**, which is now the **first node**

---

## Why returning `head` would be WRONG ❌

`head` still points to:

```
1 → None
```

So if you returned `head`, you’d lose the rest of the list.

---

## Why `prev` is the correct return value ✅

After reversing:

* **The last node becomes the new head**
* `prev` always tracks the last processed node

So:

```python
return prev
```

means:

> “Return the new head of the reversed list”

---

## Why LeetCode shows the reversed list

LeetCode internally does something like:

```python
new_head = reverseList(old_head)
print_list(new_head)
```

It doesn’t care about `prev` or `current` —
it only cares about the **node you return**.

---

## Mental model (very important)

Think like this:

* Linked list = **chain of nodes**
* The list is identified by its **head**
* Reversing changes **who the head is**
* `prev` becomes the new head

---

## One-line intuition

> **`prev` is returned because it ends up pointing to the first node of the reversed list.**
