409. Longest Palindrome

Given a string s which consists of lowercase or uppercase letters, return the length of the longest

 that can be built with those letters.

Letters are case sensitive, for example, "Aa" is not considered a palindrome.

```python
class Solution:
    def longestPalindrome(self, s: str) -> int:
        start = 0
        maxlen = 1
        for i in range(1, len(s)):
            #for even case
            l = i-1
            r = i
            while l >= 0 and r < len(s) and s[l] == s[r]:
                if r-l+1 > maxlen:
                    maxlen = r-l+1
                    start = l
                l -= 1
                r += 1
            #for odd case
            l = i-1
            r = i+1
            while l >=0 and r < len(s) and s[l] == s[r]:
                if r-l+1 > maxlen:
                    maxlen = r-l+1
                    start = l
                l -= 1
                r += 1

        return s[start:start+maxlen] # remember to return after the loop (indentation)
        
        
        
```

```python
s = "abccccdd"
```

```python
longestPalindrome(s)
```

What your code is doing

Your code solves Longest Palindromic Substring using the expand-around-center technique.

Key characteristics of your algorithm:

You fix a center (i)

Expand left and right

Check symmetry (s[l] == s[r])

Track maximum length

This is a two-pointer expansion, not binary search.

Why this is NOT binary search
Binary Search requires:

Sorted data

Monotonic condition (true → false or false → true)

Repeatedly halving the search space

❌ Your code:

Works on an unsorted string

Checks all centers

Does linear expansion, not halving

So it does not satisfy binary search properties.

Time complexity comparison

| DP Palindrome        | O(n²)           |
| Manacher’s Algorithm | O(n)            |
