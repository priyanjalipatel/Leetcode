## Correct approach for â€œlongest palindrome that can be builtâ€

### Key idea

1. Count frequency of each character
2. Use **all even counts**
3. Use **at most one odd-count character** in the center
4. Build left + center + reversed(left)

## Modified solution (returns `"dccaccd"`)

```python
from collections import Counter

class Solution:
    def longestPalindrome(self, s: str) -> str:
        freq = Counter(s)

        left = []
        center = ""

        for ch, count in freq.items():
            # add half of even counts to left
            left.append(ch * (count // 2))

            # choose ONE odd character as center
            if count % 2 == 1 and center == "":
                center = ch

        left_part = "".join(left)
        right_part = left_part[::-1]

        return left_part + center + right_part
```

````python
class Solution:
    def longestPalindrome(self, s: str) -> int:
        hm = {}
        for c in s:
            if c in hm:
                hm[c] += 1
            else:
                hm[c] = 1

        ans = 0
        isFirstOdd = False

        for key in hm:
            if hm[key] % 2 == 0:
                ans += hm[key]
            else:
                ans += hm[key]
                if not isFirstOdd:
                    isFirstOdd = True
                else:
                    ans -= 1

        return ans
````
