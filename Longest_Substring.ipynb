Because:

"abc" has no adjacent duplicates

But your code only checks neighbors, not the whole substring

Correct idea (Sliding Window)

You need to:

Keep a window of unique characters

Expand it when characters are unique

Shrink it when a duplicate appears

```python
def lengthOfLongestSubstring(self, s: str) -> int:
        charset = set()
        l = 0
        res = 0
        for r in range(len(s)):
            while s[r] in charset:
                charset.remove(s[l])
                l += 1
            charset.add(s[r])
            res = max(res, r -l + 1)
        return res
```
