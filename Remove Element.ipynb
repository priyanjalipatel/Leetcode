```python
def remove_element(num, val):
    count = 0 
    j = 0
    for i in range(0,len(num)):
        if num[i] !=val:       
            num[j] = num[i]
            j +=1
    count = len(num[:j])
    return count
```

```python
nums = [1,2,3,3,4,6,3,4]
val = 3
remove_element(nums,val)
```

```python
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        j = 0
        for i in range(0,len(nums)):
            if nums[i] !=val:
                nums[j] = nums[i]
                j +=1
        return nums

#Problem is in first value
```
